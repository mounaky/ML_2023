```python
import matplotlib.pyplot as plt
import numpy as np

#Potting x and y points

xpoint = np.array([1,2])
ypoint = np.array([4,5])
plt.plot(xpoint,ypoint)
plt.show()
```

```python
xpoint = np.array([1,2])
ypoint = np.array([4,5])
plt.plot(xpoint,ypoint,marker= 'o') #dot at the ends
  
 # plt.plot(xpoint,ypoint,'d:r')
```

```python
plt.plot(xpoint,ypoint,'d:b') #diamond shape , red color
```

```python
plt.plot(xpoint,ypoint,'D--m',ms = 10)
plt.show()
```

```python
plt.plot(xpoint,ypoint,ls='--')  #ls = line style
plt.show()
```

```python
plt.plot(xpoint,ypoint,'o:r', lw='10')  #ls = line style
plt.show()
```

```python
x=np.array([80,90,30,20,100,120,145,160,150])
y=np.array([20,30,40,50,60,100,120,130,140])
plt.plot(x,y)


 #add labels
plt.title('activity data')
plt.xlabel('averge data')
plt.ylabel('calories burnt')
plt.show()

plt.grid()
```

```python

#plot a graph where x varies frm 0 to 5 , y=x+3
#give x label as 'x-label' , y label as 'y-label', title as 'y=x+3'

x=np.array([0,1,2,3,4,5])
y=x+3
plt.plot(x,y)

plt.title('y=y+3')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()

```
